```python
df1 = df.drop('Species' , axis = 1)
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df1.values[:,1:]
X = np.nan_to_num(X)
Clustered_ds = StandardScaler().fit_transform(X)
Clustered_ds
```

```python
from sklearn.cluster import KMeans
```

```python
Clus_Num = 3
k_means = KMeans(init = "k-means++", n_clusters = Clus_Num, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
df1["Clus_km"] = labels
```

```python
df1.head()
```

```python
df1.groupby('Clus_km').mean()
```
